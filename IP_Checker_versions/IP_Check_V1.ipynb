```python
import json
import requests
import pandas as pd
```

```python
# Api key is stored in config.json file, as apikey. This simply opens the json file, then sets the key itself to the variable api_key for use later in the api calls

with open("config.json") as secret_file:
    secrets = json.load(secret_file)

api_key = secrets["VTkey"]
```

```python
ip_addys =["192.169.69.25", "99.174.190.231"]
```

```python
def ipCheck(ip_addys):
    headers = {"x-apikey" : api_key}
    url = "https://www.virustotal.com/api/v3/ip_addresses/"

    dick_list = {}
    for ip in ip_addys:
        dick_list["{0}_DF".format(ip)] = pd.DataFrame()

    return dick_list
```

```python
ipCheck(ip_addys)
```

```python
#api request params
headers = {"x-apikey" : api_key}

url = "https://www.virustotal.com/api/v3/ip_addresses/"

ipAdd = "192.169.69.25"
```

```python
# v1 api request
ip_info = requests.get(f"https://www.virustotal.com/api/v3/ip_addresses/{ipAdd}", headers=headers).json()
```

```python
# v2 api request
ip_info = requests.get("https://www.virustotal.com/api/v3/ip_addresses/" + ipAdd, headers=headers).json()
```

```python
# v3 api request
ip_info = requests.get(url + ipAdd, headers=headers).json()
```

```python
# funtcion to display json with a little bit of formatting- not super necessary
def jprint(obj):
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)

jprint(ip_info)
```
